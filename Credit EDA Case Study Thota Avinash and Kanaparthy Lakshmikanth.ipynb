### 3.2 Checking the best metric to impute the missing values

1. For CATAGORICAL DATA

```python
# I. CODE_GENDER

appl_data['CODE_GENDER'].value_counts()
```

```python
## There are some columns which the value is mentioned as 'XNA' that means 'Not Available'.
# So we have to impute Null/XNA values by suitable method(Mode).
```

```python
# II. ORGANIZATION_DATA

appl_data['ORGANIZATION_TYPE'].value_counts()
```

2. For NUMERICAL DATA

```python
# I. CNT_FAM_MEMBERS

appl_data['CNT_FAM_MEMBERS'].describe()
```

```python
### by describing the column of 'CNT_FAM_MEMBERS' it is observed that minimum value is 1.0 and maximum value is 20.0 
### Hence the maximum value is out layer. So, we have to impute null values by median method.
```

```python
appl_data['CNT_FAM_MEMBERS'].median()
```

```python
appl_data['CNT_FAM_MEMBERS']= appl_data['CNT_FAM_MEMBERS'].fillna((appl_data['CNT_FAM_MEMBERS'].median()))
```

```python
appl_data['CNT_FAM_MEMBERS'].isnull().sum()
```

```python
# II. DAYS_LAST_PHONE_CHANGE

appl_data['DAYS_LAST_PHONE_CHANGE'].describe()
```

```python
### by describing the column of 'DAYS_LAST_PHONE_CHANGE' is having the values in negative sign and 
## it is convert them into positive sign
```

```python
appl_data.loc[:, "DAYS_LAST_PHONE_CHANGE"] = appl_data.loc[:, 'DAYS_LAST_PHONE_CHANGE'].abs()
appl_data.head(2)
```

```python
appl_data['DAYS_LAST_PHONE_CHANGE'].describe()
```

```python
appl_data['DAYS_LAST_PHONE_CHANGE'].isnull().sum()
```

```python
### After converting the data type it is observed that the values between minimum and maximum are uniformely distributed
## so we have imputed null values by mean method
```

```python
appl_data['DAYS_LAST_PHONE_CHANGE']= appl_data['DAYS_LAST_PHONE_CHANGE'].fillna((appl_data['DAYS_LAST_PHONE_CHANGE'].mean()))
```
